# Structural Simplification in Popular Music

### Project Overview
This project explores whether commercially successful songs tend to be shorter, lyrically simpler, and more repetitive than the broader population of Spotify songs. It combines Billboard chart data, Spotify audio features, and song lyrics to investigate the relationship between chart success and structural simplification in popular music.

// To be expanded

### Main research question
Do commercially successful songs tend to be shorter, simpler, and more repetitive than non-charting songs?

### Hypotheses
- **H0**: There is no difference in average song duration between Billboard-charting songs and non-charting Spotify songs.
- **H1**: Billboard-charting songs are shorter on average than non-charting Spotify songs.
- **H2**: Billboard-charting songs have more repetitive lyrics than non-charting Spotify songs.
- **H3**: Billboard-charting songs have lower lexical diversity than non-charting Spotify songs.
- **H4**: Shorter duration, higher repetition, and lower lexical diversity are associated with chart success.

```python
spotify_en = spotify[(spotify['language'] == 'en') & (spotify['lyrics'].notna())].copy()
spotify_en.shape
```

```python
spotify_en[['track_name', 'track_artist']].duplicated().sum()
```

```python
spotify_en_unique = spotify_en.drop_duplicates(subset=['track_name', 'track_artist']).copy()
spotify_en_unique.shape
```

```python
spotify_en_unique[['track_name', 'track_artist']].duplicated().sum()
```

```python
billboard_song_level = billboard_charts[['song', 'artist']].drop_duplicates().copy()
billboard_song_level.shape
```

```python
spotify_en_unique['track_name_norm'] = spotify_en_unique['track_name'].str.lower()
spotify_en_unique['track_artist_norm'] = spotify_en_unique['track_artist'].str.lower()
```

```python
spotify_en_unique['track_name_norm'] = spotify_en_unique['track_name_norm'].str.replace(r"\(.*?\)", "", regex=True)
spotify_en_unique['track_name_norm'] = spotify_en_unique['track_name_norm'].str.replace(r"\[.*?\]", "", regex=True)
spotify_en_unique['track_name_norm'] = spotify_en_unique['track_name_norm'].str.replace(r"feat\.?.*", "", regex=True)
spotify_en_unique['track_name_norm'] = spotify_en_unique['track_name_norm'].str.replace(r"featuring.*", "", regex=True)
spotify_en_unique['track_name_norm'] = spotify_en_unique['track_name_norm'].str.replace(r"ft\.?.*", "", regex=True)
spotify_en_unique['track_name_norm'] = spotify_en_unique['track_name_norm'].str.replace(r"[^a-z0-9\s]", "", regex=True)
spotify_en_unique['track_name_norm'] = spotify_en_unique['track_name_norm'].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
spotify_en_unique['track_artist_norm'] = spotify_en_unique['track_artist_norm'].str.replace(r"\(.*?\)", "", regex=True)
spotify_en_unique['track_artist_norm'] = spotify_en_unique['track_artist_norm'].str.replace(r"\[.*?\]", "", regex=True)
spotify_en_unique['track_artist_norm'] = spotify_en_unique['track_artist_norm'].str.replace(r"feat\.?.*", "", regex=True)
spotify_en_unique['track_artist_norm'] = spotify_en_unique['track_artist_norm'].str.replace(r"featuring.*", "", regex=True)
spotify_en_unique['track_artist_norm'] = spotify_en_unique['track_artist_norm'].str.replace(r"ft\.?.*", "", regex=True)
spotify_en_unique['track_artist_norm'] = spotify_en_unique['track_artist_norm'].str.replace("&", "and", regex=False)
spotify_en_unique['track_artist_norm'] = spotify_en_unique['track_artist_norm'].str.replace(r"[^a-z0-9\s]", "", regex=True)
spotify_en_unique['track_artist_norm'] = spotify_en_unique['track_artist_norm'].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
billboard_song_level['song_norm'] = billboard_song_level['song'].str.lower()
billboard_song_level['artist_norm'] = billboard_song_level['artist'].str.lower()
```

```python
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"\(.*?\)", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"\[.*?\]", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"feat\.?.*", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"featuring.*", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"ft\.?.*", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"[^a-z0-9\s]", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"\(.*?\)", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"\[.*?\]", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"feat\.?.*", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"featuring.*", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"ft\.?.*", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace("&", "and", regex=False)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"[^a-z0-9\s]", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
spotify_billboard_match = spotify_en_unique.merge(
    billboard_song_level,
    left_on=['track_name_norm', 'track_artist_norm'],
    right_on=['song_norm', 'artist_norm'],
    how='left'
)
```

```python
spotify_billboard_match['song'].notna().sum()
```

```python
spotify_billboard_match['is_charting'] = spotify_billboard_match['song'].notna().astype(int)
spotify_billboard_match['is_charting'].value_counts()
```

```python
matched = spotify_billboard_match['is_charting'].sum()
total = len(spotify_billboard_match)
print("Matched songs:", matched)
print("Total songs:", total)
print("Match percentage:", round(matched / total * 100, 2), "%")
```

```python
spotify_billboard_match[spotify_billboard_match['is_charting'] == 1][
    ['track_name', 'track_artist', 'song', 'artist']
].head(20)
```

After restricting the Spotify dataset to English-language songs with available lyrics and removing repeated playlist-based entries, 13,734 unique song-artist observations remained. A first exact-match merge with Billboard chart data, using normalized song and artist names, identified 3,063 songs that appeared in Billboard at least once. This corresponds to approximately 22.3% of the cleaned Spotify sample. The result suggests that the merged dataset is suitable for comparing charting and non-charting songs.
