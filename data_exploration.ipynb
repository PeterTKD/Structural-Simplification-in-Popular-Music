# Data Preparation

In this stage, the Billboard dataset is aggregated from weekly chart observations to song level. This allows each Spotify song to be matched against a single Billboard song record. Song and artist names are normalized in both datasets in order to reduce mismatches caused by punctuation, featured-artist notation, and formatting differences. The merged dataset then includes both a binary charting label `is_charting` and richer success indicators such as peak rank, weeks on board, and chart entries.

```python
billboard_song_level = (
    billboard_charts.groupby(['song', 'artist'], as_index=False)
    .agg({
        'rank': ['min', 'mean'],
        'peak-rank': 'min',
        'weeks-on-board': 'max',
        'date': ['min', 'max', 'count']
    })
)
```

```python
billboard_song_level.columns = [
    'song',
    'artist',
    'best_weekly_rank',
    'avg_weekly_rank',
    'peak_rank',
    'weeks_on_board',
    'first_chart_date',
    'last_chart_date',
    'chart_entries'
]
```

```python
billboard_song_level.head()
```

```python
billboard_song_level.shape
```

```python
billboard_song_level['song_norm'] = billboard_song_level['song'].str.lower()
billboard_song_level['artist_norm'] = billboard_song_level['artist'].str.lower()

billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"\(.*?\)", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"\[.*?\]", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"feat\.?.*", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"featuring.*", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"ft\.?.*", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"[^a-z0-9\s]", "", regex=True)
billboard_song_level['song_norm'] = billboard_song_level['song_norm'].str.replace(r"\s+", " ", regex=True).str.strip()

billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"\(.*?\)", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"\[.*?\]", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"feat\.?.*", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"featuring.*", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"ft\.?.*", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace("&", "and", regex=False)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"[^a-z0-9\s]", "", regex=True)
billboard_song_level['artist_norm'] = billboard_song_level['artist_norm'].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
spotify_billboard_match = spotify_en_unique.merge(
    billboard_song_level,
    left_on=['track_name_norm', 'track_artist_norm'],
    right_on=['song_norm', 'artist_norm'],
    how='left'
)
```

```python
spotify_billboard_match['is_charting'] = spotify_billboard_match['peak_rank'].notna().astype(int)
```

```python
spotify_billboard_match['weeks_on_board'] = spotify_billboard_match['weeks_on_board'].fillna(0)
spotify_billboard_match['chart_entries'] = spotify_billboard_match['chart_entries'].fillna(0)
```

```python
spotify_billboard_match = spotify_billboard_match.drop(
    columns=['playlist_name', 'playlist_id', 'playlist_genre', 'playlist_subgenre'],
    errors='ignore'
)
```
